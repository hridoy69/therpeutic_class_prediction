# MACCS FP Fingerprint Calculation

```python
from rdkit import Chem
from rdkit.Chem import MACCSkeys
```

```python
def generate_MACCS_FP_from_SMILES(smiles):
    # Create a molecule object from a SMILES string
    mol = Chem.MolFromSmiles(smiles)
    
    # Calculate the MACCS keys fingerprint
    maccs_fp = MACCSkeys.GenMACCSKeys(mol)
    maccs_fp_list = maccs_fp.ToList()
    return maccs_fp_list
```

```python
df_filtered.columns
```

```python
# Step 1: Generate MACCS fingerprints for all SMILES strings
maccs_fp_list = df_filtered['SMILES'].apply(generate_MACCS_FP_from_SMILES)

# Step 2: Convert the list of fingerprints into a DataFrame
maccs_fp_df = pd.DataFrame(
    maccs_fp_list.tolist(),
    columns=[f'MACCS_Bit_{i}' for i in range(1, len(maccs_fp_list.iloc[0]) + 1)],
    index=df_filtered.index
)

# Step 3: Combine original DataFrame with MACCS fingerprints
maccs_fp_df_filtered = pd.concat([df_filtered, maccs_fp_df], axis=1)

# Step 4: Ensure the desired column order (optional)
desired_columns = ['ChEMBL ID', 'Target/Cell lines', 'Phytochemicals', 'Inchikeys',
                   'Diseases', 'Inchi', 'SMILES', 'Target'] + [f'MACCS_Bit_{i}' for i in range(1, len(maccs_fp_list.iloc[0]) + 1)]
maccs_fp_df_filtered = maccs_fp_df_filtered[desired_columns]
maccs_fp_df_filtered
```
